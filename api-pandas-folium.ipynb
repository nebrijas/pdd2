```python
import pandas as pd
import folium
```

```python
url_zrgz = 'https://www.zaragoza.es/sede/servicio/transporte/accidentalidad-trafico/accidente.csv'
coord_zrgz = [41.649693,-0.887712]
```

```python
mapa = folium.Map(location=coord_zrgz)
mapa
```

```python
df = pd.read_csv(url_zrgz,delimiter=";")
df
```

```python
longitudes = []
latitudes = []
```

```python
longitudes
```

```python
type(longitudes)
```

```python
for i in df['geometry']:
    punto_coord = i.split(',')
    longitudes += [float(punto_coord[1])]
longitudes
```

```python
for i in df['geometry']:
    punto_coord = i.split(',')
    latitudes += [float(punto_coord[0])]
latitudes
```

```python
df_coord = pd.DataFrame({'long':longitudes,'lat':latitudes})
df_coord
```

```python
df_accidentes = pd.concat([df['type'],df_coord],axis=1)
df_accidentes
```

```python
type(df_accidentes['lat'][0])
```

```python
marcador = folium.Marker(coord_zrgz, popup="¡Hola, Zaragoza!")
mapa = mapa.add_child(marcador)
mapa
```

```python
for index, fila in df_accidentes.iterrows():
    marcador = folium.Marker([fila['lat'],fila['long']],\
                            popup=fila['type'])
    mapa.add_child(marcador)
mapa
```
